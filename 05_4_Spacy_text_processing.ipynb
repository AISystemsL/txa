```python
import requests
url = "https://en.wikipedia.org/wiki/Pisa"
html = requests.get(url).content.decode()
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, "html5lib")
text = '\n'.join([''.join(s.findAll(text=True))for s in soup.findAll('p')])
print(text)
```

```python
import spacy.cli
import IPython.core.ultratb
import sys

try:
    spacy.cli.download('en_core_web_sm')
    # spacy.cli.download('en')
    # spacy.cli.download('it')
except Exception as exc:
    tb = IPython.core.ultratb.VerboseTB()
    print(tb.text(*sys.exc_info()))
```

```python
import spacy
#nlp = spacy.load('en')
nlp = spacy.load('en_core_web_sm')
##nlp = spacy.load('en')
#nlp = spacy.load('de')
#nlp = spacy.load('es')
#nlp = spacy.load('pt')
#nlp = spacy.load('fr')
#nlp = spacy.load('it')
#nlp = spacy.load('nl')
#nlp = spacy.load('xx')
```

```python
doc = nlp(text)
```

```python
doc
```

```python
for i, sentence in enumerate(doc.sents):
    print(i,'-',sentence.text)
```

```python
print('text', 'lemma', 'pos', 'tag', 'dep',
          'shape', 'alpha', 'stop', sep='\t')
print()
test_sent = list(doc.sents)[24]
for token in test_sent:
    print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
          token.shape_, token.is_alpha, token.is_stop, sep='\t')
```

### chunking

```python
list(nlp('She sells sea shells by the sea shore').noun_chunks)
```

```python
list(doc.noun_chunks)
```

### entities

```python
list(doc.ents)
```

```python
for ent in doc.ents:
    if ent.label_=="GPE":
        print(ent.text, ent.start_char, ent.end_char, ent.label_)
```

```python
from spacy import displacy

displacy.render(doc, style='ent', jupyter=True)
```

```python
displacy.render(nlp('Andrea Esuli is a researcher at the Italian National Research Council. He lives in Italy.'), style='ent', jupyter=True)
```

### parsing

```python
displacy.render(nlp('I like to play with Python.'), style='dep', jupyter=True)
```

```python
print('text', 'lemma', 'pos', 'tag', 'dep',
          'shape', 'alpha', 'stop', 'anchestor', sep='\t')
print()
for token in test_sent:
    print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
          token.shape_, token.is_alpha, token.is_stop, next(token.ancestors,'ROOT'), sep='\t')
```

```python
displacy.render(test_sent, style='dep', jupyter=True)
```

### Other languages

```python
nlp_it = spacy.load('en_core_web_sm')  # spacy.load('it')
doc_it = nlp_it('Andrea Esuli lavora al Consiglio Nazionale delle Ricerche e vive a Pisa.')
displacy.render(doc_it, style='ent', jupyter=True)
displacy.render(doc_it, style='dep', jupyter=True)
```

```python
doc_it = nlp_it('Maria gli disse di ascoltarla, ma lui non la guardava.')
for token in doc_it:
    print(token, token.tag_, sep='\t')
```

```python
print('text', 'lemma', 'pos', 'tag', 'dep',
          'shape', 'alpha', 'stop', 'anchestor', sep='\t')
print()
for token in doc_it:
    print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
          token.shape_, token.is_alpha, token.is_stop, next(token.ancestors,'ROOT'), sep='\t')
```
